```python
import os
import nltk
import json
import pandas as pd
from path import Path
%matplotlib inline
from pandas.io.json import json_normalize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
nltk.download('vader_lexicon')
```

```python
# Fetch top news articles about oil
nasdaq_headlines = newsapi.get_everything(q="Nasdaq", language="en")
nasdaq_headlines
```

```python
# Fetch news about the financial crisis on 2008 in English
nasdaq_news_en = newsapi.get_everything(
    q="nasdaq",
    language="en"
)
#save json
with open('nasdaq_news_en.json', 'w') as json_file:
    json.dump(nasdaq_news_en, json_file)


```

```python
#read in the json file NDAQ
nasdaq_news_en_df = pd.read_json ('nasdaq_news_en.json')
#parse the json and load into a dataframe
nasdaq_news_en_df = json_normalize(nasdaq_news_en_df['articles'])
#drop all columns except for description
nasdaq_news_en_df = nasdaq_news_en_df.drop(['author', 'title', 'url','urlToImage','publishedAt','content','source.id','source.name'], axis=1)
#instantiate vader sentiment 
sentiment_title = nasdaq_news_en_df.apply(lambda x : analyzer.polarity_scores(x['description']), axis=1)
#add a column for each sentimement metric
nasdaq_news_en_df['Comp'] =sentiment_title.apply(lambda x : x.get('compound') )
nasdaq_news_en_df['Pos'] =sentiment_title.apply(lambda x : x.get('pos') )
nasdaq_news_en_df['Neu'] =sentiment_title.apply(lambda x : x.get('neu') )
nasdaq_news_en_df['Neg'] =sentiment_title.apply(lambda x : x.get('neg') )
nasdaq_news_en_df
```
